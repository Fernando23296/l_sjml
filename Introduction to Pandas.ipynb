# Introduccion a Pandas
Pandas es una API de analisis de datos en columnas, por lo cual es ideal para manipular y analizar datos de entrada. Un plus para su uso es la disponibilidad en muchos frameworks de ML , que lo aceptan como entrada. 

```python
import pandas as pd
pd.__version__
```

Las estructuras de datos principales de pandas estan implementadas en dos clases:<br>
    <b>DataFrame</b>: es como un tabla de datos relacional, con filas y columnas con nombre.<br>
    <b>Series</b>: es una columna simple. Una clase DataFrame incluye una o mas <b>Series</b> y un nombre para cada <b>Series</b><br><br>
    El <b>DataFrame</b> es una abstraccion que se usa para manipular datos.<br><br>
    La manera de crear una <b>Series</b> es de la siguiente manera:

```python
pd.Series(["La Paz","Cochabamba","Tarija"])
```

Los objetos de DataFrame pueden crearse haciendo uso de un diccionario, donde se asignan los nombres de las columnas con los de las Series que se desean "meter". Si estas Series no coinciden en longitud, los valores que falten se complementan con valores NA/NaN.

```python
nombre_ciudades = pd.Series(["la paz","santa cruz","oruro"])
poblacion = pd.Series([22222,33333,11233])

pd.DataFrame({"Ciudades":nombre_ciudades,"Poblacion":poblacion})
```
